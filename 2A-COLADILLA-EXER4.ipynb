```python
import pandas as pd
import numpy as np #numerical operation
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data_path = "Lesson_4_student_grades_dataset.csv"
dataset = pd.read_csv(data_path)
statistics = dataset.describe()

x = dataset['Hours_of_Study']
x1 = dataset['Attendance_Percentage']
x2 = dataset['Previous_Exam_Scores']
x3 = dataset['Extracurricular_Activities']
x4 = dataset['Travel_Time_to_School']
y = dataset['Grades']

print(dataset.isnull().sum())
```

```python
x = dataset.drop('Grades', axis = 1)
y = dataset['Grades']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

model = LinearRegression()

model.fit(x_train,y_train)
```

```python
new_data = np.array([[5,90,80,1,30]])
prediction = model.predict(new_data)

print(f"Predicted Grade: {prediction[0]}")
```
